```python
import re
from collections import Counter

# Utilizamos una expresión regular para encontrar todas las palabras
# que están entre comillas o sin comillas en cada fila de 'amenities'
df_clean['amenities_list'] = df_clean['amenities'].apply(lambda x: re.findall(r'"(.*?)"|(\w+)', x))

# Concatenamos todas las palabras en una lista grande
all_amenities = [amenity[0] if amenity[0] else amenity[1] for sublist in df_clean['amenities_list'] for amenity in sublist]

# Contamos la frecuencia de cada palabra
amenities_counts = Counter(all_amenities)

# Convertimos el resultado en un DataFrame para mejor visualización
amenities_df = pd.DataFrame(list(amenities_counts.items()), columns=['Servicio', 'Frecuencia']).sort_values(by='Frecuencia', ascending=False)

# Imprimimos el DataFrame con las frecuencias de los servicios
amenities_df
```

```python
plt.Figure(figsize=(10,6))
sns.barplot(x='Servicio',y='Frecuencia',data=amenities_df[amenities_df['Frecuencia']>1000])
plt.xlabel('Servicios')
plt.ylabel('Frecuencia')
plt.title('Servicios mas populares')
plt.xticks(rotation=90)
plt.show()
```
